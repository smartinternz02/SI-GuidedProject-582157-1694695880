```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dff=sns.load_dataset('car_crashes')
```

```python
dff.head(6)
```

```python
dff.tail(8)
```

```python
dff.info()
```

```python
sns.scatterplot(x="total",y="alcohol",data=dff)
```

```python
Inference from the plot is as car crashes increases as alcohol consumption increases.
Directly proportional
```

```python
sns.scatterplot(x="speeding",y="total",data=dff)
```

```python
Inference:This means that as the speed increases,the total number of car crashes also increases.
The trend line shows that the relationship is not linear, but rather exponential.
```

```python
sns.scatterplot(x="not_distracted",y="total",data=dff)
```

```python
Inference:This means that as the level of distraction increases,the number of car accidents also increases
```

```python
Inference from the graph:-as car crashes increases as alcohol consumption increases.
```
